```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the data
df['late_encoded'] = label_encoder.fit_transform(df['late'])

print(df)
```

#### Point estimate/ Sample statistic
* Single value derived from a sample of data, used to estimate a corresponding population parameter.
* The key idea is that a point estimate provides a best guess for an unknown population parameter based on sample data.

For example: If you count the number of defective items in a small sample from a production line, this count is a point estimate of the number of defective items in the entire production.


Calculating the sample mean

```python
# Calculate the proportion of late shipments
late_prop_samp = (df['late'] == "Yes").mean()

# Print the results
print("The proportion of late shipments is:", late_prop_samp * 100, "%")
```

Calculating a z-score

Bootstrap Distribution

```python
late_shipments_boot_distn = []
for i in range(1000):  # Number of bootstrap iterations
    sample = df.sample(frac=1, replace=True)['late_encoded']
    mean = np.mean(sample)
    late_shipments_boot_distn.append(mean)
```

Variables can have different ranges and units, so we need to standardize them to make comparisons meaningful. For instance, a hypothesis test that changes results based on whether values are in Euros or US dollars wouldn't be very useful.

* Key standardized value: z-score. 
* To find it: sample statistic, the hypothesized statistic, and the standard error (which is estimated from the bootstrap distribution).

```python
# Hypothesize that the proportion is 6%
late_prop_hyp = 0.06

# Calculate the standard error
std_error = np.std(late_shipments_boot_distn)

# Find z-score of late_prop_samp
z_score = (late_prop_samp - late_prop_hyp)/std_error

# Print z_score
print(z_score)
```

Calculating p-value

To decide between the null hypothesis and the alternative hypothesis, calculate the p-value from the z-score.
Using the late shipments dataset, the null hypothesis (Ho) is that the proportion of late shipments is 6%. The alternative hypothesis (Ha) is that the proportion is greater than 6%.

```python
# Calculate the p-value
p_value = 1 - norm.cdf(z_score)
                 
# Print the p-value
print(p_value) 
```

Calculating a confidence interval

A single estimate of a sample statistic is rarely exact. For instance, if the hypothesized proportion of late shipments is 6%, the actual proportion in any new sample will likely differ slightly due to sampling variability. Therefore, it's better to provide a confidence interval, such as saying, "We are 95% confident that the proportion of late shipments is between A and B.

```python
# Calculate 95% confidence interval using quantile method
lower = np.quantile(late_shipments_boot_distn, 0.025)
upper = np.quantile(late_shipments_boot_distn, 0.975)

# Print the confidence interval
print(lower, upper)
```

Does the confidence interval match up with the conclusion to stick with the original assumption that 6% is a reasonable value for the unknown population parameter?

Yes, since 0.06 is included in the 95% confidence interval and we failed to reject H0 due to a large p-value, the results are similar.
