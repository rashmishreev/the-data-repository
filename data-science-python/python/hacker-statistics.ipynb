### Problem Statement:

A player is attempting to ascend the Empire State Building by playing a dice game. The rules are as follows:

- Rolling a 1 or 2 results in one step down.
- Rolling a 3, 4, or 5 results in one step up.
- Rolling a 6 allows the player to roll again and ascend the resulting number of steps.
- The player cannot go below step 0.
- There is a 0.1% chance of falling down and starting over at step 0 after any move.
- The player has made a bet with a friend that they can reach step 60.

Determine the probability of the player winning the bet.

```python
# Import numpy and set seed
import numpy as np
np.random.seed(123)

# Use randint() to simulate a dice
print(np.random.randint(1, 7))

# Use randint() again
print(np.random.randint(1, 7))
```

```python
# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1, 7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <= 5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice, step)
```

This code simulates a random walk, inspired by the concept of a 
player moving up the Empire State Building based on dice rolls.

In a random walk, each step is determined by a random event. 
This is analogous to the path of a molecule in a fluid or the 
fluctuating fortunes of a gambler. 

This simulation will utilize a 'for' loop to record the player's 
position after each step in the random walk. -->

```python
# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = max(0, step -1)       # The player cannot go below step 0.
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt


# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

Simulate multiple walks

```python
# This code simulates multiple random walks to estimate the probability 
# of reaching the 60th step in the Empire State Building game. 

# A single random walk simulation provides only one outcome. 
# By repeatedly simulating the walk and collecting the results, 
# we can gain insights into the likelihood of reaching the 60th step. 
# This approach helps us assess the player's chances of winning the bet.
```

```python
# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk five times
for i in range(5) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

```python
# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
# To account for the player's clumsiness, a random chance of falling 
# down is introduced. 

# A random float between 0 and 1 is generated. 
# If this value is less than or equal to 0.005 (0.5% probability), 
# the player falls down and the step counter is reset to 0.
```

```python
# Simulate random walk 20 times
all_walks = []
for i in range(20) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001:
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

```python
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001:
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

```python
# To calculate the odds of reaching at least 60 steps, we analyze the 'ends' array. 

# 'ends' is a NumPy array containing the final positions reached in 500 simulated random walks. 

# The probability of reaching 60 steps or higher is estimated by:
# 1. Counting the number of elements in 'ends' that are 
#    greater than or equal to 60.
# 2. Dividing this count by the total number of simulations (500).

# This calculation provides an estimate of the player's chances of winning the Empire State Building game.
```

```python
ends_size = ends.size   # Size of ends numpy array
chance = ends >= 60     # Boolean value of Integers in 'ends' that are >=60
print(ends[chance])     # Elements in 'ends' that are >=60
ends60_size = ends[chance].size     #Number of integers that are greater than or equal to 60
print("\nNumber of integers in ends array that are greater than or equal to 60 is", ends60_size)
print("\nThe estimated chance that the player will reach at least 60 steps high playing the Empire State Building game is",(ends60_size/ends_size)*100,"%")
```
