## Informative and attractive visualizations in Python using the Seaborn library

**Seaborn**
- Seaborn is a powerful Python library that makes it easy to create informative and attractive data visualizations.
- Data visualization is often a huge component of both the data exploration phase and the communication of results, so Seaborn is very useful.
-  We can use Seaborn to create a variety of plots, including scatter plots, count plots, bar plots, and box plots, and can also customize visualizations.


**Advantages**
- Seaborn's main purpose is to make data visualization easy. It was built to automatically handle a lot of complexity behind the scenes.
- Works extremely well with pandas data structures. pandas is a Python library that is widely used for data analysis.
- It's built on top of Matplotlib, which is another Python visualization library. Matplotlib is extremely flexible. Seaborn allows to take advantage of this flexibility when needed, while avoiding the complexity that Matplotlib's flexibility can introduce.

**Importing Seaborn Library**
- The line "`import seaborn as sns`" will import Seaborn as the conventionally used alias "`sns`". 
- "`sns`"?: The Seaborn library was apparently named after a character named *Samuel Norman Seaborn* from the television show "The West Wing" - thus, the standard alias is the character's initials. 
- Also need to import Matplotlib, which is the library that Seaborn is built on top of. "`import matplotlib.pyplot as plt`". "`plt`" is the alias that most people use to refer to Matplotlib, so we'll use that here as well.

````python
# Converting GDP ($ per capita), Literacy (%) and Phones (per 1000) into a list
gdp = countries_of_world['GDP ($ per capita)'].to_list()
````

````python
# Literacy (%) column should be a float datatype
countries_of_world["Literacy (%)"] = (
    countries_of_world["Literacy (%)"]
    .astype(str)  # Convert to string to handle any unexpected types
    .str.replace(",", ".")  # Replace "," with "." (if needed)
    .astype(float)  # Convert to float
)
````

````python
# Phones (per 1000) column should be a float datatype
countries_of_world["Phones (per 1000)"] = (
    countries_of_world["Phones (per 1000)"]
    .astype(str)  # Convert to string to handle any unexpected types
    .str.replace(",", ".")  # Replace "," with "." (if needed)
    .astype(float)  # Convert to float
)
````

````python
phones = countries_of_world['Phones (per 1000)'].tolist()
percent_literate = countries_of_world['Literacy (%)'].tolist()
````

````python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)
plt.show()
````

````python
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=percent_literate)
# Show plot
plt.show()
````

While this plot does not show a linear relationship between GDP and percent literate, countries with a lower GDP do seem more likely to have a lower percent of the population that can read and write.

````python
# Create count plot with region on the y-axis
sns.countplot(y=countries_of_world["Region"])  # Correct way to reference the column
# Show plot
plt.show()
````

````python
# Responses to a survey sent out to young people. Primary question was: how many young people surveyed report being scared of spiders? 
# Survey participants were asked to agree or disagree with the statement "I am afraid of spiders". 
# Responses vary from 1 to 5, where 1 is "Strongly disagree" and 5 is "Strongly agree".

# Create a count plot with "Spiders" on the x-axis
sns.countplot(x="Spiders", data=survey_responses)

# Display the plot
plt.show()
````

````python
# Create a scatter plot of absences vs. final grade
sns.scatterplot(x="absences", y="G3", 
                data=student_alchohol_consumption, 
                hue="location",
                hue_order = ['Rural', 'Urban'])

# Show plot
plt.show()
````

It looks like students with higher absences tend to have lower grades in both rural and urban areas.

````python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x='school', data=student_alchohol_consumption, hue='location', palette=palette_colors)

# Display plot
plt.show()
````

Students at GP tend to come from an urban location, but students at MS are more evenly split.

## Visualizing Two Quantitative Variables

````python
# Using relplot for scatter plot to arrange the plots in rows/columns and separate charts
sns.relplot(x="absences", y="G3", 
            data=student_alchohol_consumption,
            kind="scatter", 
            col="study_time")

# Show plot
plt.show()
````

````python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_alchohol_consumption,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row='famsup',
            row_order=["yes", "no"])

# Show plot
plt.show()
````

It looks like the first semester grade does correlate with the final grade, regardless of what kind of support the student received.

Seaborn's mpg dataset, which contains one row per car model and includes information such as the year the car was made, the number of miles per gallon ("M.P.G.") it achieves, the power of its engine (measured in "horsepower"), and its country of origin.

- What is the relationship between the power of a car's engine ("horsepower") and its fuel efficiency ("mpg")? 
- And how does this relationship vary by the number of cylinders ("cylinders") the car has?

````python
mpg = sns.load_dataset('mpg')
mpg.head(5)
````

````python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower",
            y="mpg", 
            data=mpg,
            kind="scatter", 
            size="cylinders",
            hue='cylinders')

# Show plot
plt.show()
````

Cars with higher horsepower tend to get a lower number of miles per gallon. They also tend to have a higher number of cylinders.

````python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(kind='scatter',data=mpg,x='acceleration',y='mpg',hue='origin',style='origin')
# Show plot
plt.show()
````

Cars from the USA tend to accelerate more quickly and get lower miles per gallon compared to cars from Europe and Japan.
